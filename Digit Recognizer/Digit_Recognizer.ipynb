```python
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(10,6)
matplotlib.rcParams['figure.facecolor']='#00000000'
```

```python
def ret_x():
    x=train_data.iloc[:,1:]
    return x.to_numpy()

def ret_y():
    y=train_data['label']
    return y.to_numpy()
```

```python
def plot_digits(no_of_rows,no_of_columns):
    fig,axs=plt.subplots(no_of_rows,no_of_columns,figsize=(12,12))
    index=0
    x,y=ret_x(),ret_y()
    for i in range(no_of_rows):
        for j in range(no_of_columns):
            axs[i,j].imshow(x[index].reshape(28,28),cmap='binary')
            axs[i, j].set_xticks([])
            axs[i, j].set_yticks([])
            axs[i, j].text(x=24, y=4, s=str(y[index]), c='blue', size=12)
            index += 1
```

```python
plot_digits(8,8)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler=MinMaxScaler(feature_range=(0,1))
```

```python
help(MinMaxScaler)
```

```python
x=ret_x()
min_max_scaler.fit(x)
```

```python
min_max_scaler.data_max_
```

```python
min_max_scaler.data_min_
```

```python
x_scaled=min_max_scaler.transform(x)
```

```python
x_scaled.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
help(LogisticRegression)
```

```python
model=LogisticRegression(solver='saga', max_iter=1000)
```

```python
y=ret_y()
model.fit(x_scaled,y)
```

```python
import joblib
```

```python
objs={
    "scaler":min_max_scaler,
    "model":model
}
```

```python
joblib.dump(objs,"objs.sav")
```
