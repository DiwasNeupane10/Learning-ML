```python
def max_depth_error(md):
    model=DecisionTreeClassifier(max_depth=md,random_state=42)
    model.fit(X_train,train_target_df)
    train_error=1-model.score(X_train,train_target_df)
    val_error=1-model.score(X_val,val_target_df)
    return {'Max_depth':md,'Training_error':train_error,'Validation_error':val_error}
```

```python
error_df=pd.DataFrame([max_depth_error(md) for md in range(1,21)])
```

```python
error_df
```

***From the graph we can see that the model can be regularized by tuning the hyperparameter max_depth at 7 which elimnates overfitting***

```python
def max_depth_and_max_leaf_nodes_error(md,n):
    model=DecisionTreeClassifier(max_depth=md,max_leaf_nodes=n,random_state=42)
    model.fit(X_train,train_target_df)
    train_error=1-model.score(X_train,train_target_df)
    val_error=1-model.score(X_val,val_target_df)
    return{'Max-depth':md,'Max_leaf_nodes':n,'Training-error':train_error,'Validation-error':val_error}
    
```

```python
%%time
errors_df=pd.DataFrame([max_depth_and_max_leaf_nodes_error(md,n) for md in range(1,21) for n in range(2,100)])
```
