```python
RFM_date=pd.Timestamp('2011-12-11 10:15:00')
```

```python
RFM_date
```

```python
x=merged_df.groupby('CustomerID').agg({'Recency': lambda date: (RFM_date - date)})
```

```python
x.rename(columns={"Recency":"RecencyDays"},inplace=True)
```

```python
x
```

```python
merged_df=pd.merge(merged_df,x,on='CustomerID')
```

```python
merged_df
```

```python
merged_df['RecencyScore']=pd.qcut(merged_df['RecencyDays'],q=5,labels=[5, 4, 3, 2, 1])
```

```python
merged_df
```

```python
merged_df['FrequencyScore']=pd.qcut(merged_df['Frequency'].rank(method='first'),q=5,labels=[1, 2, 3, 4, 5])
```

```python
merged_df
```

```python
merged_df['MonetaryScore']=pd.qcut(merged_df['TotalPrice'].rank(method="first"),q=5,labels=[1,2,3,4,5])
```

```python
merged_df
```

```python
merged_df['Segments']=merged_df['RecencyScore'].astype(str)+merged_df['FrequencyScore'].astype(str)+merged_df['MonetaryScore'].astype(str)
```

```python
merged_df
```

```python
segments_dict={
    r'5[4-5]':'Champions',
    r'[3-4][4-5]':'Loyal Customers',
    r'[4-5][2-3]':'Potential Loyalists',
    r'51':'New Customers',
    r'41':'Promising',
    r'3[1-2]':'Need Attention',
    r'33':'About to Sleep',
    r'[1-2]5':'Cannot Lose them',
    r'[1-2][3-4]':'At Risk',
    r'[1-2][1-2]':'Hibernating'
}
```

```python
merged_df['fm'] = ((merged_df['FrequencyScore'].astype(int) + merged_df['MonetaryScore'].astype(int))/2).round().clip(1,5).astype(int)
```

```python
merged_df
```

```python
merged_df['CustomerSegments']=merged_df['RecencyScore'].astype(str)+merged_df['fm'].astype(str)
```

```python
merged_df.drop(columns=['Segments'],axis=1,inplace=True)
```

```python
merged_df
```

```python
merged_df['CustomerSegments']=merged_df['CustomerSegments'].replace(segments_dict,regex=True)
```

```python
merged_df
```

```python
RFM=merged_df[['CustomerID','CustomerSegments','TotalPrice']]
```
